## two Gaussian distributions 

Figure 3.3 plots the pdf of a Gaussian-distributed random variable $x$ for two different variances. In both
cases, the mean of the random variable is fixed at $\bar{x}=10$
while the variance is $\sigma_x^2=100$ in one case and
$\sigma_x^2=10$ in the other. We explain in chpater 4 that the pdf of a Gaussian random variable is defined in terms of $(\bar{x},\sigma_x^2)$ by the following expression (see expression 4.4): 


$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad f_x(x) = \frac{1}{\sqrt{2\pi}\sigma_x}e^{-(x-\bar{x})^2/2\sigma_x^2},\quad x\in(-\infty,\infty) \qquad (\mbox{Gaussian})$



From Fig.3.3 we observe that the smaller the
variance of $x$ is, the more concentrated its pdf is around its mean.

```python
mean = 10
var1 = 100
var2 = 10

a1 =1/np.sqrt(2*np.pi*var1)
a2 =1/np.sqrt(2*np.pi*var2)

x = np.arange(-40, 60, 0.1)
f1 = a1 * np.exp(-1/2*(x-mean)**2/var1)
f2 = a2 * np.exp(-1/2*(x-mean)**2/var2)

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

plt.figure(figsize=(11,3));
plt.plot(x, f1)
plt.xlabel('$x$')
plt.ylabel('$f_x(x)$')
plt.title('two Gaussian distribution');
plt.annotate('$\sigma_x^2 = 100$',xy=(20,0.03),xytext=(25,0.04),arrowprops={"width":1,"headwidth":3,'headlength':7})
plt.grid(visible = True, alpha =0.3);
plt.xlim([-40, 60])
plt.ylim([0, 0.05])

plt.figure(figsize=(11,3));
plt.plot(x, f2)
plt.xlabel('$x$')
plt.ylabel('$f_x(x)$')
plt.annotate('$\sigma_x^2 = 10$',xy=(15,0.1),xytext=(25,0.13),arrowprops={"width":1,"headwidth":3,'headlength':7})
plt.grid(visible = True, alpha =0.3);
plt.xlim([-40, 60])
plt.ylim([0, 0.15]);


```

## two Rayleigh distributions

Figure 3.4 provides similar plots for a second random
variable $x$ with a Rayleigh distribution, namely, with a pdf given by 

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad f_x(x) = \frac{x}{\alpha^2}e^{-x^2/2\alpha^2},\quad x\geq 0,\quad \alpha >0 \qquad (\mbox{Rayleigh})$ 

where $\alpha$ is a
positive parameter. The value of $\alpha$ determines the mean and variance of $x$
according to the following expressions (see Prob. 3.15): 

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \bar{x} = \alpha \sqrt{\frac{\pi}{2}},\;\;\;\;\;\;\;\;\sigma_x^2=\left(2-\frac{\pi}{2}\right)\alpha^2
\label{eq1.1.4}$ 

so that, in contrast to the Gaussian case,
the mean and variance of a Rayleigh-distributed random variable
cannot be chosen independently of each other since their values are linked
through $\alpha$. In Fig. 3.4, the
top plot corresponds to $\bar{x}=3.7599$ and $\sigma_x^2=3.8628$, while
the bottom plot corresponds to $\bar{x}=1.2533$ and $\sigma_x^2=0.4292$.

```python
alpha1 = 3
alpha2 = 1

x = np.arange(0.01, 12, 0.1)
g1 = (x/(alpha1**2))*np.exp(-x**2/(2*alpha1**2))
g2 = (x/(alpha2**2))*np.exp(-x**2/(2*alpha2**2));

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

plt.figure(figsize=(11,3));
plt.plot(x, g1)
plt.xlabel('$x$')
plt.ylabel('$f_x(x)$')
plt.title('two Rayleigh distributions');
plt.annotate('$\sigma_x^2 = 3.8628$',xy=(5,0.15),xytext=(6,0.2),arrowprops={"width":1,"headwidth":3,'headlength':7})
plt.grid(visible = True, alpha =0.3);
plt.xlim([0, 12])
plt.ylim([0, 0.25])

plt.figure(figsize=(11,3));
plt.plot(x, g2)
plt.xlabel('$x$')
plt.ylabel('$f_x(x)$')
plt.annotate('$\sigma_x^2 = 0.4292$',xy=(2,0.4),xytext=(3,0.5),arrowprops={"width":1,"headwidth":3,'headlength':7})
plt.grid(visible = True, alpha =0.3);
plt.xlim([0, 12])
plt.ylim([0, 0.7])


c = 2-(np.pi/2);
var1  = c*(alpha1**2)
var2  = c*(alpha2)**2
mean1 = alpha1*np.sqrt(np.pi/2)
mean2 = alpha2*np.sqrt(np.pi/2)
print("mean and variance of the first distribution: ",mean1, var1)
print("mean and variance of the second distribution: ",mean2, var2)
```

## plot of tanh function

Another conditional mean computation

Assume $x$ is a binary random variable that assumes the values $\pm 1$ with probability $1/2$ each. Assume in addition that we observe a noisy realization of $x$, say,

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  y = x + v$

where $v$ is a zero-mean Gaussian random variable that is independent of $x$ and has variance equal to 1, i.e., its pdf is given by

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  f_v(v) = \frac{1}{\sqrt{2\pi}}\;e^{-v^{2}/2}$


Let us evaluate the conditional mean of $x$ given observations of $y$. From definition (3.100), we find that we need to know the conditional
pdf, $f_{x|y}(x|y)$, in order to evaluate the integral expression. For this purpose, we call upon future result
(3.160), which states that the pdf of the sum of two independent random variables, namely, $y=x+v$, is equal to the convolution of their individual pdfs, i.e.,

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad f_y(y)=\int_{-\infty}^{\infty}f_x(x)f_v(y-x)dx$

In this example we have

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad f_x(x) = \frac{1}{2}\delta(x-1)\;+\;\frac{1}{2}\delta(x+1)$

where $\delta(\cdot)$ is the Dirac delta function, so that
$f_y(y)$ is given by

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad f_y(y) = \frac{1}{2}f_v(y+1)+\frac{1}{2}f_v(y-1)$

Moreover, in view of the Bayes rule (3.42b), the joint pdf of $\{x,y\}$ is given by

$\qquad \qquad \qquad \qquad \qquad \qquad \begin{eqnarray} f_{x,y}(x,y) = f_x(x)\, f_{y|x}(y|x)\\
&=&(\frac{1}{2}\delta(x-1)\;+\;\frac{1}{2}\delta(x+1))\,
f_{v}(y-x)\\
&=&\frac{1}{2}f_{v}(y -1)\delta(x-1)+\frac{1}{2}f_{v}(y +1)
\delta(x+1)\end{eqnarray}$

Using the Bayes rule again we get

$\qquad \qquad \qquad \qquad \qquad \qquad f_{x|y}(x|y)=\frac{f_{x,y}(x,y)}{f_{y}(y)}$

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad=\frac{f_{v}(y-1)\delta(x-1)}{f_{v}(y +1)+f_{v}(y -1)} + \frac{f_{v}(y+1)\delta(x+1)}{f_{v}(y +1)+f_{v}(y -1)}$

Substituting  into expression (3.100) and integrating we obtain

$\begin{eqnarray}
\qquad \qquad \qquad \qquad \qquad \quad E(x|y)&=&\frac{f_{v}(y -1)}{f_{v}(y +1)+f_{v}(y
-1)}-\frac{f_{v}(y +1)}
{f_{v}(y +1)+f_{v}(y -1)}\\&&\\
&=& \dfrac{1}{\left(\frac{e^{-(y +1)^{2}/2}}{e^{-(y
-1)^{2}/2}}\right)+1} - \frac{1}{\left(\frac{e^{-(y
-1)^{2}/2}}{e^{-(y +1)^{2}/2}}\right)+1}\\
&=&\dfrac{e^y-e^{-y}}{e^y+e^{-y}}\\ &=&\mbox{tanh}(y)
\end{eqnarray}$

In other words, the conditional mean of $x$ given observations of $y$ is the
hyperbolic tangent function, which is shown in Fig. 3.5.


```python
x = np.arange(-3, 13, 0.1)
h = np.tanh(x)

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

plt.figure(figsize=(11,3));
plt.plot(x, h)
plt.xlabel('$y$')
plt.ylabel('$tanh(y)$')
plt.grid(visible = True, alpha =0.3);
plt.xlim([-3, 3])
plt.ylim([-1, 1]);
x = [0]
y = [0]
plt.plot(x, y, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="red")
line = plt.vlines(0, -1, -0.05, linestyle="dashed", linewidth=1)
line.set_color("black")
line = plt.hlines(0, -3, -0.05, linestyle="dashed", linewidth=1)
line.set_color("black")
```
